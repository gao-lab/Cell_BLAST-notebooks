```python
import os
import numpy as np
import Cell_BLAST as cb
import utils
cb.config.RANDOM_SEED = 0
cb.config.N_JOBS = 4
fixed_model_kwargs = dict(
    latent_dim=10, cat_dim=20,
    epoch=500, patience=20
)
```

# Mouse
### Quake_Smart-seq2_Large_Intestine

```python
quake_smart_seq2_large_intestine = cb.data.ExprDataSet.read_dataset("../../Datasets/data/Quake_Smart-seq2_Large_Intestine/data.h5")
utils.peek(quake_smart_seq2_large_intestine, "Quake_Smart-seq2_Large_Intestine")
quake_smart_seq2_large_intestine = quake_smart_seq2_large_intestine.normalize()
quake_smart_seq2_large_intestine.obs.head()
```

```python
opt_model_kwargs = dict(batch_effect="donor")
quake_smart_seq2_large_intestine_model = cb.directi.fit_DIRECTi(
    quake_smart_seq2_large_intestine, quake_smart_seq2_large_intestine.uns["seurat_genes"],
    **fixed_model_kwargs, **opt_model_kwargs,
    path="/tmp/cb/db/Quake_Smart-seq2_Large_Intestine/model_0"
)
quake_smart_seq2_large_intestine.latent = quake_smart_seq2_large_intestine_model.inference(quake_smart_seq2_large_intestine)
```

```python
plot = quake_smart_seq2_large_intestine.visualize_latent("cell_type1")
plot.get_figure().savefig("Quake_Smart-seq2_Large_Intestine/cell_type1.pdf", bbox_inches="tight")
```

```python
plot = quake_smart_seq2_large_intestine.visualize_latent("donor")
plot.get_figure().savefig("Quake_Smart-seq2_Large_Intestine/donor.pdf", bbox_inches="tight")
```

```python
plot = quake_smart_seq2_large_intestine.visualize_latent("gender")
plot.get_figure().savefig("Quake_Smart-seq2_Large_Intestine/gender.pdf", bbox_inches="tight")
```

```python
%%capture capio
quake_smart_seq2_large_intestine_models = [quake_smart_seq2_large_intestine_model]
opt_model_kwargs = dict(batch_effect="donor")
for i in range(1, cb.config.N_JOBS):
    print("==== Model %d ====" % i)
    quake_smart_seq2_large_intestine_models.append(cb.directi.fit_DIRECTi(
        quake_smart_seq2_large_intestine, quake_smart_seq2_large_intestine.uns["seurat_genes"],
        **fixed_model_kwargs, **opt_model_kwargs,
        random_seed=i, path="/tmp/cb/db/Quake_Smart-seq2_Large_Intestine/model_%d" % i
    ))
quake_smart_seq2_large_intestine_blast = cb.blast.BLAST(quake_smart_seq2_large_intestine_models, quake_smart_seq2_large_intestine, 
                                               keep_exprs=True).build_empirical()
quake_smart_seq2_large_intestine_blast.save("Quake_Smart-seq2_Large_Intestine")
```
